```python
def timMang(mang, text):
    if len(text) <= 0: return None
    for i in range(len(mang)):
        if len(''.join(mang[i:])) <= len(text):
            return mang[i:]

def timMangSau(mang, text):    
    if len(text) <= 0: return None
    for i in range(len(mang)):
        if len(''.join(mang[:i])) == len(text):
            return mang[:i]
        if len(''.join(mang[:i])) >= len(text):
            return mang[:i-1]
```

```python
f = open('textTT2Fail.txt')
textTT2Fail = json.load(f)

f1 = open('otextTT2Fail.txt')
otextTT2Fail = json.load(f1)     

for i in FileFail:
    if textTT2Fail[i] != None:
        check = getText1(textTT2Fail[i])
        if check == False:
            textTT2Fail[i] = None
            otextTT2Fail[i] = None
        else:
            textTT2Fail[i] = check[1]
            otextTT2Fail[i] = timMang(otextTT2Fail[i], check[1])
            
RaEoOAFail = []
toxicokineticsFail = []
for i in FileFail:
    if textTT2Fail[i] != None:        
        check = getData1(textTT2Fail[i])
        if check != False:
            if check[1] == 1: toxicokineticsFail.append(i)
            RaEoOAFail.append(' '.join(timMangSau(otextTT2Fail[i], check[0][0])))
            textTT2Fail[i] = check[0][1]
            otextTT2Fail[i] = timMang(otextTT2Fail[i], check[0][1])
        else:
            RaEoOAFail.append('ERROR: cannot get data')
            textTT2Fail[i] = 'ERROR: cannot get data'
            otextTT2Fail[i] = 'ERROR: cannot get data'
    else:
        RaEoOAFail.append(None)
        
for i in toxicokineticsFail:
    check = getText2(textTT2Fail[i])
    if check != False:
        textTT2Fail[i] = check[1]
        otextTT2Fail[i] = timMang(otextTT2Fail[i], check[1])
        
distributionFail = []
for i in FileFail:
    if textTT2Fail[i] != None and textTT2Fail[i] != 'ERROR: cannot get data':
        check = getData2(textTT2Fail[i])
        if check != False:
            mang = timMangSau(otextTT2Fail[i], check[0])
            if mang != None:
                distributionFail.append(' '.join(mang))
            else:
                distributionFail.append(None)
            textTT2Fail[i] = check[1]            
            otextTT2Fail[i] = timMang(otextTT2Fail[i], check[1])     
        else:
            distributionFail.append('ERROR: cannot get data')
            textTT2Fail[i] = 'ERROR: cannot get data'            
            otextTT2Fail[i] = 'ERROR: cannot get data'  
    elif textTT2Fail[i] == None:
        distributionFail.append(None)
    else:
        distributionFail.append('ERROR: cannot get data')
        
potentialFail = []

for i in FileFail:
    if textTT2Fail[i] != None and textTT2Fail[i] != 'ERROR: cannot get data':
        check = getData3(textTT2Fail[i])
        if check != False:
            mang = timMangSau(otextTT2Fail[i], check[0])
            if mang != None:
                potentialFail.append(' '.join(mang))
            else:
                potentialFail.append(None)
            textTT2Fail[i] = check[1]            
            otextTT2Fail[i] = timMang(otextTT2Fail[i], check[1])
        else:
            potentialFail.append('ERROR: cannot get data')
            textTT2Fail[i] = 'ERROR: cannot get data'            
            otextTT2Fail[i] = 'ERROR: cannot get data'  
    elif textTT2Fail[i] == None:
        potentialFail.append(None)
    else:
        potentialFail.append('ERROR: cannot get data')
        
RaEoEFail = []

for i in FileFail:
    if textTT2Fail[i] != None and textTT2Fail[i] != 'ERROR: cannot get data':
        check = getData4(textTT2Fail[i])
        if check != False:
            mang = timMangSau(otextTT2Fail[i], check[0])
            if mang != None:
                RaEoEFail.append(' '.join(mang))
            else:
                RaEoEFail.append(None)
            textTT2Fail[i] = check[1]            
            otextTT2Fail[i] = timMang(otextTT2Fail[i], check[1])
        else:
            RaEoEFail.append('ERROR: cannot get data')
            textTT2Fail[i] = 'ERROR: cannot get data'            
            otextTT2Fail[i] = 'ERROR: cannot get data' 
    elif textTT2Fail[i] == None:
        RaEoEFail.append(None)
    else:
        RaEoEFail.append('ERROR: cannot get data')
        
metaboliumFail = []

for i in FileFail:
    if textTT2Fail[i] != None and textTT2Fail[i] != 'ERROR: cannot get data':
        check = getData5(textTT2Fail[i])
        if check != False:
            mang = timMangSau(otextTT2Fail[i], check[0])
            if mang != None:
                metaboliumFail.append(' '.join(mang))
            else:
                metaboliumFail.append(None)
            textTT2Fail[i] = check[1]            
            otextTT2Fail[i] = timMang(otextTT2Fail[i], check[1])
        else:
            metaboliumFail.append('ERROR: cannot get data')
            textTT2Fail[i] = 'ERROR: cannot get data'            
            otextTT2Fail[i] = 'ERROR: cannot get data' 
    elif textTT2Fail[i] == None:
        metaboliumFail.append(None)
    else:
        metaboliumFail.append('ERROR: cannot get data')

toxidicFail = dict()
otoxidicFail = dict()

for i in FileFail:
    if textTT2Fail[i] != None and textTT2Fail[i] != 'ERROR: cannot get data':
        check = getToxi(textTT2Fail[i])
        if check != False:
            toxidicFail[i] = check[0]
            otoxidicFail[i] = timMangSau(otextTT2Fail[i], check[0])
            textTT2Fail[i] = check[1]
            otextTT2Fail[i] = timMang(otextTT2Fail[i], check[1])
        else:
            toxidicFail[i] = textTT2Fail[i]
            otoxidicFail[i] = otextTT2Fail[i]
    elif textTT2Fail[i] == None:
        toxidicFail[i] = None
        otoxidicFail[i] = None
    else:
        toxidicFail[i] = 'ERROR: cannot get data'
        otoxidicFail[i] = 'ERROR: cannot get data'

toxi1Fail = []
toxi2Fail = []

for i in toxidicFail:
    if toxidicFail[i] != None and toxidicFail[i] != 'ERROR: cannot get data':
        check = getData6(toxidicFail[i])
        if check == False:
            toxi1Fail.append(' '.join(otoxidicFail[i]))
            toxi2Fail.append(None)
        else:
            mang1 = timMangSau(otoxidicFail[i], check[0])
            mang2 = timMang(otoxidicFail[i], check[1])
            if mang1 != None:
                toxi1Fail.append(' '.join(mang1))
            else:
                toxi1Fail.append(None)  
                
            if mang2 != None:
                toxi2Fail.append(' '.join(mang2))
            else:
                toxi2Fail.append(None)
    elif toxidicFail[i] == None:
        toxi1Fail.append(None)
        toxi2Fail.append(None)
    else:
        toxi1Fail.append('ERROR: cannot get data')
        toxi2Fail.append('ERROR: cannot get data')        
        
LD50_OralFail = []

for i in FileFail:
    if textTT2Fail[i] != None and textTT2Fail[i] != 'ERROR: cannot get data':
        check = getData7(textTT2Fail[i])
        if check != False:
            LD50_OralFail.append(' '.join(timMangSau(otextTT2Fail[i], check[0])))
            textTT2Fail[i] = check[1]
            otextTT2Fail[i] = timMang(otextTT2Fail[i], check[1])
        else:
            LD50_OralFail.append('ERROR: cannot get data')
            textTT2Fail[i] = 'ERROR: cannot get data'            
            otextTT2Fail[i] = 'ERROR: cannot get data' 
    elif textTT2Fail[i] == None:
        LD50_OralFail.append(None)
    else:
        LD50_OralFail.append('ERROR: cannot get data')
        

LD50_DermalFail = []

for i in FileFail:
    if textTT2Fail[i] != None and textTT2Fail[i] != 'ERROR: cannot get data':
        check = getData8(textTT2Fail[i])
        if check != False:
            mang = timMangSau(otextTT2Fail[i], check[0])
            if mang == None:
                LD50_DermalFail.append(None)
            else:
                LD50_DermalFail.append(' '.join(mang))
            textTT2Fail[i] = check[1]
            otextTT2Fail[i] = timMang(otextTT2Fail[i], check[1])
        else:
            LD50_DermalFail.append('ERROR: cannot get data')
            textTT2Fail[i] = 'ERROR: cannot get data'            
            otextTT2Fail[i] = 'ERROR: cannot get data'
    elif textTT2Fail[i] == None:
        LD50_DermalFail.append(None)
    else:
        LD50_DermalFail.append('ERROR: cannot get data')
        
        
LD50_InhalationFail = []

for i in FileFail:
    if textTT2Fail[i] != None and textTT2Fail[i] != 'ERROR: cannot get data':
        check = getData9(textTT2Fail[i])
        if check == False:
            LD50_InhalationFail.append('ERROR: cannot get data')
            textTT2Fail[i] = 'ERROR: cannot get data'
            otextTT2Fail[i] = 'ERROR: cannot get data'
        else:          
            mang = timMangSau(otextTT2Fail[i], check[0])
            if mang == None:
                LD50_InhalationFail.append(None)
            else:
                LD50_InhalationFail.append(' '.join(mang))
            textTT2Fail[i] = check[1]
            otextTT2Fail[i] = timMang(otextTT2Fail[i], check[1])
    elif textTT2Fail[i] == None:
        LD50_InhalationFail.append(None)
    else:
        LD50_InhalationFail.append('ERROR: cannot get data')    
        
        
skin_irritationFail = []

for i in FileFail:
    if textTT2Fail[i] != None and textTT2Fail[i] != 'ERROR: cannot get data':
        check = getData10(textTT2Fail[i])
        if check != False:
            mang = timMangSau(otextTT2Fail[i], check[0])            
            if mang == None:
                skin_irritationFail.append(None)
            else:
                skin_irritationFail.append(' '.join(mang))
            textTT2Fail[i] = check[1]
            otextTT2Fail[i] = timMang(otextTT2Fail[i], check[1])
        else:
            skin_irritationFail.append('ERROR: cannot get data')
            textTT2Fail[i] = 'ERROR: cannot get data'
            otextTT2Fail[i] = 'ERROR: cannot get data'
    elif textTT2Fail[i] == None:
        skin_irritationFail.append(None)
    else:
        skin_irritationFail.append('ERROR: cannot get data')
          

eye_irritationFail = []

for i in FileFail:
    if textTT2Fail[i] != None and textTT2Fail[i] != 'ERROR: cannot get data':
        check = getData11(textTT2Fail[i])
        if check == False:
            eye_irritationFail.append(' '.join(otextTT2Fail[i]))
        else:  
            mang = timMangSau(otextTT2Fail[i], check[0]) 
            if mang == None:
                eye_irritationFail.append(None)
            else:
                eye_irritationFail.append(' '.join(mang))
    elif textTT2Fail[i] == None:
        eye_irritationFail.append(None)
    else:
        eye_irritationFail.append('ERROR: cannot get data')
        
```
