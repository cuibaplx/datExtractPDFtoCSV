```python
#tim mang otext để gắn dữ liệu về trạng thái ban đầu

def timMang(mang, text):
    if len(text) <= 0: return None
    for i in range(len(mang)):
        if len(''.join(mang[i:])) <= len(text):
            return mang[i:]

def timMangSau(mang, text):    
    if len(text) <= 0: return None
    for i in range(len(mang)):
        if len(''.join(mang[:i])) == len(text):
            return mang[:i]
        if len(''.join(mang[:i])) >= len(text):
            return mang[:i-1]
```
